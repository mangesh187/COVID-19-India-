```
# hw to plot data from model
m1 = stats.norm(9,.6)
l1 = m1.rvs(size=1000)
```

```
plt.hist(l1,linewidth=.2,bins=10,edgecolor='black')
```

```
#mean , kurtosis, iqr, mode , skewmess,geometric mean , harmonic mean####
```

```
import scipy.stats
```

```
gmean = stats.gmean(l1,axis=0)
kurtosis = stats.kurtosis(l1,axis=0)
iqr = stats.iqr(l1,axis=0)
median = stats.median_absolute_deviation(l1,axis=0)
skewd = stats.skew(l1,axis=0)
hmean = stats.hmean(l1,axis=0)
mode = stats.mode(l1,axis=0)
```

```
print(gmean,kurtosis,iqr,median, skewd,hmean,mode)
      
      
     
      
```
